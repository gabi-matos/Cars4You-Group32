```python
def incoherencies(df, dictionary):
    """
    Correct inconsistencies by converting negative values to positive and capping values above a threshold
    """
    for key, value in dictionary.items():
        if value == 0:
            n = (df[key] < 0).sum()
            if n > 0:
                print(f'{key} has {n} negative values made absolute.')
            df.loc[df[key] < 0, key] = df.loc[df[key] < 0, key].abs()
        else:
            n = (df[key] > value).sum()
            if n > 0:
                print(f'{key} has {n} values above {value}, capped at {value}.')
            df.loc[df[key] > value, key] = value
    return df
```

```python
def floats_to_int(df, variables):
    """
    Convert float values to integers and return the number of conversions
    """
    for var in variables:
        sum_floats = df[var].apply(lambda x: isinstance(x, float)).sum() 
        df[var] = df[var].round().astype('Int64')
        print(f'{var} has {sum_floats} that were converted to integers')
    return df
```

```python
def outliers_capping(dataset, limits):
    """
    Cap numerical values to specified upper or lower limits to handle outliers
    """
    df=dataset.copy()
    for key, value in limits.items():
        if key != 'year':
            df[key] = df[key].clip(upper = value)
        else:
            df[key] = df[key].clip(lower = value) 
    return df
```

```python
def price_capping(df, low, high):
    """
    Receives the dataframe and its limits and clips the target 
    accordingly - returning it.
    """
    df_cap = df.clip(low, high)
    
    return df_cap
```

```python
def feature_engineering(df):
    """
    Create new features from existing columns to enhance predictive power
    """
    #create carAge
    df['carAge'] = 2020 - df['year']
    
    #create mileagePerYear
    df['mileagePerYear'] = np.where(
    (df['carAge'].notna()) & (df['carAge'] != 0) ,
    df['mileage'] / df['carAge'],
    1)
    
    #create mpg_per_EngineSize
    df['mpg_per_engineSize'] = np.where(
    (df['engineSize'].notna()) & (df['engineSize'] != 0),
    df['mpg'] / df['engineSize'],
    0)
    
    #create carAge*Mileage
    df['carAge*Mileage'] = df['carAge'] * df['mileage']
    
    # log(mileage)
    df['log(mileage)'] = np.log1p(df["mileage"])
    
    # create a binary variable, 1 for luxury brands (Audi, Mercedes & BMW), 0 for non luxury
    df['is_luxury'] = df['Brand'].isin(['BMW', 'Audi', 'Mercedes']).astype(int)
    
    return df
```

```python
def fill_categorical_unknown(df, lista):
    """
    Fill missing values in specified categorical columns with 'Unknown'
    """
    for col in lista:
        df[col] = df[col].fillna('Unknown')
    return df
```

```python
def scale_training(df, features):
    """
    Scale specified features in the training dataset and return the scaled DataFrame and the fitted scaler.
    """
    #create scaler
    scaler = StandardScaler() # best MAE test performance on kaggle
    #scaler = MinMaxScaler() #- performance on test data slightly better with StandardScaler() - increased MAE by around 4
    #scaler = RobustScaler() #- performance on test data slightly better with StandardScaler() - increased MAE by around 3
    
    #fit transform to training data
    df_scalled_array = scaler.fit_transform(df[features])
    
    #transform into dataframe
    df_train_scaled = pd.DataFrame(df_scalled_array, columns=features, index=df.index)
    
    return df_train_scaled, scaler
```

```python
def fill_MV(df, variables_knn, imputer, medians):
    """
    Fill missing values in the dataset using a KNN imputer, medians, and 0 for 'previousOwners'
    """
    df_scaled_mv = df.copy()
    
    #apply KNN imputer to variables: 'mileage', 'tax', 'mpg', 'engineSize', 'mileagePerYear', 'mpg_per_engineSize'
    df_scaled_mv[variables_knn] = imputer.transform(df_scaled_mv[variables_knn])
    
    #fill year, carAge and carAge*Mileage with median
    for key, value in medians.items():
        df_scaled_mv[key] = df_scaled_mv[key].fillna(value)
    
    #fill PreviousOwners and hasDamage with 0
    df_scaled_mv['previousOwners'] = df_scaled_mv['previousOwners'].fillna(0)
    
    return df_scaled_mv
```

```python
def run_models(x_train, y_train, x_val, y_val, model):
    """
    Takes the training X, Y and validation X and Y dataframes and a model
    Returns the MAE of both training and validation
    """
    
    #fitting the model to training data
    model.fit(x_train, y_train)
    
    #predicting on training and transforming back to euros from log
    pred_train = model.predict(x_train)
    
    #calculating training MAE
    mae_train = mean_absolute_error(y_train, pred_train)
    
    #prediction on validation and transforming back to euros from log
    pred_val = model.predict(x_val)
    
    #calculating validation MAE
    mae_val = mean_absolute_error(y_val, pred_val)
    
    #print mae
    print(f'MAE train {model}: {mae_train}\nMAE val {model}: {mae_val}')
    
    return mae_train, mae_val
```

```python
def get_best_model(x, y, ps, model, grid, n=80):
    """
    Takes X, Y, the predefined split, a model and a parameter grid and uses the algorithm Randomized Search to return
    the best combination of parameters and the validation MAE that it gets.
    """
    model_search = RandomizedSearchCV(estimator=model, param_distributions=grid, n_iter=n,
                                     scoring='neg_mean_absolute_error', cv=ps, random_state=42,
                                     n_jobs=-1, verbose=1)
    
    model_search.fit(x,y)
    best_params = model_search.best_params_
    mae = -model_search.best_score_
    
    return best_params, mae
```

```python
def error_by_price_bin(model, x_val, y_val, x_train, y_train, name):
    """
    Creates bins for price and segments the MAE in each segment"""
    
    model.fit(x_train, y_train)
    y_pred = model.predict(x_val)
    df = pd.DataFrame({
        'y_true': y_val,
        'y_pred': y_pred
    })
    df['abs_error'] = (df['y_true'] - df['y_pred']).abs()
    df['price_bin'] = pd.qcut(df['y_true'], 5)
    out = df.groupby('price_bin')['abs_error'].mean()
    out.name = name
    return out
```

```python
##create function to blend 2 models

def blend_2_models(model1, model2, x_val, y_val, step=0.05):
    
    """
    Create a function that takes 2 models and a validation set (x_val and y_val) and returns the best
    weights for each model and its mean absolute error
    """
    
    #fit data
    model1.fit(x_train, y_train)
    model2.fit(x_train, y_train)
    
    #predictions in x_val
    pred1 = model1.predict(x_val)
    pred2 = model2.predict(x_val)
    
    best_mae = np.inf
    best_weights = None
    
    for w in np.arange(0.2, 0.9, 0.01):
        blend = w * pred1 + (1-w) * pred2
        mae = mean_absolute_error(y_val, blend)
        
        if mae < best_mae:
            best_mae = mae
            best_weights = w
            
    return best_weights
```

```python
## function to blend 3 models
def blend_3_models(model1, model2, model3, x_val, y_val, step=0.05):
    
    """
    Create a function that takes 3 models and a validation set (x_val and y_val) and returns the best
    weights for each model and its mean absolute error
    """
    
    #fit data
    model1.fit(x_train, y_train)
    model2.fit(x_train, y_train)
    model3.fit(x_train, y_train)
    
    #prediction in x_Val
    pred1 = model1.predict(x_val)
    pred2 = model2.predict(x_val)
    pred3 = model3.predict(x_val)
    
    best_mae = np.inf
    best_weights = None
    
    for w1 in np.arange(step, 0.25 + 1e-9, 0.02):
        for w2 in np.arange(0.25, 0.45 + 1e-9, 0.02):
            w3 = 1 - w1 - w2
            if w3 <= 0:
                continue
                
            blend = w1 * pred1 + w2 * pred2 + w3 * pred3
            mae = mean_absolute_error(y_val, blend)
            
            if mae < best_mae:
                best_mae = mae
                best_weights = (w1, w2, w3)
    
    return best_weights  
```

```python
def run_blend_2_models(model1, model2, x_train, y_train, x_val, y_val, w):
    """A function that takes 2 models, the train and validation X and Y and the weights and returns their MAE"""
    
    #fit data
    model1.fit(x_train, y_train)
    model2.fit(x_train, y_train)
    
    #predict for train
    pred1_train = model1.predict(x_train)
    pred2_train = model2.predict(x_train)
    
    # create weighted average for train
    blend_train = w * pred1_train + (1-w) * pred2_train
    
    #calculate mae for train
    mae_train = mean_absolute_error(y_train, blend_train)
    
    #predict for val
    pred1_val = model1.predict(x_val)
    pred2_val = model2.predict(x_val)
    
    # create weighted average for train
    blend_val = w * pred1_val + (1-w) * pred2_val
    
    #calculate mae for train
    mae_val = mean_absolute_error(y_val, blend_val)
    
    return mae_train, mae_val
```

```python
def run_blend_3_models(model1, model2, model3, x_train, y_train, x_val, y_val, weights):
    """A function that takes 3 models, the train and validation X and Y and the weights and returns their MAE"""
    
    #fit data
    model1.fit(x_train, y_train)
    model2.fit(x_train, y_train)
    model3.fit(x_train, y_train)
    
    #predict for train
    pred1_train = model1.predict(x_train)
    pred2_train = model2.predict(x_train)
    pred3_train = model3.predict(x_train)
    
    # create weighted average for train
    blend_train = weights[0] * pred1_train + weights[1] * pred2_train + weights[2] * pred3_train
    
    #calculate mae for train
    mae_train = mean_absolute_error(y_train, blend_train)
    
    #predict for val
    pred1_val = model1.predict(x_val)
    pred2_val = model2.predict(x_val)
    pred3_val = model3.predict(x_val)
    
    # create weighted average for train
    blend_val = weights[0] * pred1_val + weights[1] * pred2_val + weights[2] * pred3_val
    
    #calculate mae for train
    mae_val = mean_absolute_error(y_val, blend_val)
    
    return mae_train, mae_val
```

```python
def blend_2_models_predict(model1, model2, w, test, x_full, y_full):
    """
    creates a function that receives 2 models, its weights and a test dataset and returns
    predictions on the test data
    """
    #fit on full data 
    model1.fit(x_full, y_full)
    model2.fit(x_full, y_full)
    
    #predict on test data
    pred1 = model1.predict(test)
    pred2 = model2.predict(test)
    
    #do weighted average
    blend_pred = w * pred1 + (1-w) * pred2
    
    return np.maximum(blend_pred, 0)
```

```python
def blend_3_models_predict(model1, model2, model3, weights, test):
    """
    creates a function that receives 3 models, its weights and a test dataset and returns
    predictions on the test data
    """
    w1, w2, w3 = weights
    
    #fit on full data 
    model1.fit(x_full, y_full)
    model2.fit(x_full, y_full)
    model3.fit(x_full, y_full)
    
    #predict on test data
    pred1 = model1.predict(test)
    pred2 = model2.predict(test)
    pred3 = model3.predict(test)
    
    #do weighted average
    blend_pred = w1 * pred1 + w2 * pred2 + w3 * pred3
    
    predictions = pd.DataFrame({
        'carID' : test.index.astype('Int32'),
        'price' : blend_pred
    })
    
    return predictions
```

```python
def predict_test(model, x_full, y_full, test):
    """Takes the model and the full X and Y datasets and fits the model in the full dataset 
    predicts on test data and returns a dataframe with predictions"""
    
    #fit model on full data
    model.fit(x_full, y_full)
    
    #predict on test data
    predictions = model.predict(test)
    
    #create dataframe with predictions
    predictions_df = pd.DataFrame({
        'carID' : test.index.astype('Int32'),
        'price' : predictions
    })
    
    return predictions_df
```

## **7. Future Work**


We are aware that our final model overfits and still deviates around 1265.70€ from the actual price of the cars, but we believe that it is a good result for the project that was proposed. However, if we had more time to develop our work, we would probably explore Cross-Validation. This would result in a better split of our data because it is a more robust method, using multiple splits to average performance. To present a good implementation of CV, we would need more time to search and test it; therefore, we opted for the Holdout Method.

For the future, it would be essential to retrain and monitor the models' performance to ensure that predictions remain accurate and aligned with evolving market conditions, since the new data will be changing over time. For example, inflation rates, fluctuations in the economy itself, or in the car market.
